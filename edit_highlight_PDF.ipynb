```python
import pdfplumber
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text
file_path =  r"/content/LeavePolicy 1.pdf"
resulted_text = extract_text_from_pdf(file_path)
print(resulted_text)
```

```python
def replace_text_in_pdf(pdf_path, query, new_text):
    """
    Replaces the occurrences of a query string with new text in a PDF file.
    """
    doc = fitz.open(pdf_path)
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text_instances = page.search_for(query)
        for inst in text_instances:
            # Instead of delete_text which is not available, we redact the text first
            annot = page.add_redact_annot(inst)  # Add a redaction annotation
            annot.update()  # Apply the annotation
            page.apply_redactions()  # Redact the text (make it invisible)

            # Insert new text at the location of the redacted text
            # Use the top-left corner of the rectangle as the insertion point
            page.insert_text(inst.tl, new_text)  # inst.tl gives the top-left point (x0, y0)
    doc.save("modified_pdf.pdf")  # Save the modified PDF with a new name

file_path = r"/content/LeavePolicy 1.pdf"  # Replace with your PDF file path
query_string = "Subject to local laws and regulations"  # Replace with your query
new_text = "Updated text here"  # Replace with the new text

replace_text_in_pdf(file_path, query_string, new_text)
print("PDF with replaced text saved as 'modified_pdf.pdf'")
```
